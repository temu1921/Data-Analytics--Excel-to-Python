```python
xls = pd.ExcelFile('/content/Clinical QI Data Set.xlsx')
df1 = pd.read_excel(xls, 'Hypertension') #assign the first tab
df2 = pd.read_excel(xls, 'Diabetes') #assign the 2nd tab
df3 = pd.read_excel(xls, 'Obesity') #assign the 3rd tab
```

### HYPERTENSION

```python
#1st method: assigning each possible values
df1.loc[(df1['Systolic'] <140 ) & (df1['Diastolic'] < 90), 'BP Controlled'] = 'Y'
df1.loc[(df1['Systolic'] >=140 ) & (df1['Diastolic'] >= 90), 'BP Controlled'] = 'N'
df1.loc[(df1['Systolic'] >=140 ) & (df1['Diastolic'] < 90), 'BP Controlled'] = 'N'
df1.loc[(df1['Systolic'] < 140 ) & (df1['Diastolic'] >= 90), 'BP Controlled'] = 'N'
df1
```

```python
#2nd method: creating a function
def label_htension(rowS):
  if rowS['Systolic'] < 140:
    if rowS['Diastolic'] < 90:
      return 'Y'
  return 'N'

# df1.apply(label_htension, axis=1) #to check if the function works as expected
df1['BP Controlled2'] = df1.apply(label_htension, axis=1)
df1
```

```python
df2['Last_A1c'] = df2['Last_A1c'].replace("[QNS> ]","", regex=True)
df2.loc[df2['Last_A1c'].isnull(), 'Last_A1c'] = '0'
```

```python
#implement the data cleaning logic to the actual data
df2['Last_A1c'] = df2['Last_A1c'].replace("[<>,]","", regex = True) #remove the additional characters but keeping the number
df2['Last_A1c'] = df2['Last_A1c'].replace("[QNS]","0", regex = True) #replace the value to 0
print(df2['Last_A1c'].unique()) #check if the code works as expected
```

```python
df2['Last_A1c'] = pd.to_numeric(df2['Last_A1c'], errors='coerce') #convert string to numeric for calculation
```

```python
#METHOD 1
df2.loc[(df2['Last_A1c'] < 7 ) , 'Goal']= 'Controllable'
df2.loc[(df2['Last_A1c'] > 9 ) , 'Goal']= 'Uncontrollable'
df2.loc[(df2['Last_A1c'] >= 7 ) & (df2['Last_A1c'] <= 9 ) , 'Goal'] = 'Good'
df2
```

```python
#METHOD 2
#2nd method: creating a function
def label_diabetes(rowS):
  if rowS['Last_A1c'] < 7:
    return 'Controllable'
  if rowS['Last_A1c'] > 9:
      return 'Uncontrollable'
  return 'Good'

# df2.apply(label_diabetes, axis=1) #to check if the function works as expected
df2['Goal2'] = df2.apply(label_diabetes, axis=1)
```

### OBESITAS

```python
df3['BMI'] = df3['BMI'].replace("<", "")
df3['BMI'] = pd.to_numeric(df3['BMI'], errors='coerce')
```

```python
# METHOD 1
df3.loc[(df3['BMI'] < 15 ) , 'Diagnosis']= 'Underweight'
df3.loc[(df3['BMI'] >= 15 ) & (df3['BMI'] <= 25 ) , 'Diagnosis'] = 'Normal'
df3.loc[(df3['BMI'] > 25 ) & (df3['BMI'] <= 30 ) , 'Diagnosis'] = 'Overweight'
df3.loc[(df3['BMI'] > 30 ) , 'Diagnosis']= 'Obese'
df3
```

```python
# METHOD 2
def label_obesitas(rowS):
  if rowS['BMI'] < 15:
    return 'Underweight'
  if rowS['BMI'] >= 15 and rowS['BMI'] <= 25:
    return 'Normal'
  if rowS['BMI'] > 25 and rowS['BMI'] <= 30:
    return 'Overrweight'
  # if rowS['BMI'] > 30:
  return 'Obese'

# df1.apply(label_htension, axis=1) #to check if the function works as expected
df3['Diagnosis2'] = df3.apply(label_obesitas, axis=1)
df3
```
